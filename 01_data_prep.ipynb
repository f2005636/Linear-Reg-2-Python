** 1. import libraries **

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

** 2. import dataset **

```python
df = pd.read_csv('kc_house_data.csv')
df.drop(['Unnamed: 0','id','zipcode','lat','long','sqft_living','sqft_lot'], axis=1, inplace=True)
df.head()
```

** 3. year **

```python
def Y_year (x):
    if x[1] == 0: return (x[0])
    else: return (x[1])
```

```python
df['yr'] = df[['yr_built','yr_renovated']].apply(Y_year, axis=1)
df.drop(['date','yr_built','yr_renovated'], axis=1, inplace=True)
df.head()
```

** 4. export ** 

```python
df.to_csv('data_prep.csv')
```
